```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# 1. jester (rating range from -10 to 10)
df = pd.read_csv('Dataset/UserRatings1.csv')
df.head()
```

```python
df_ = df.melt(id_vars='JokeId', value_name='rating')
df_.dropna(inplace=True)
df_.variable = df_.variable.str.replace('User', '')
# df_.columns = ['itemID', 'userID', 'rating']
df_.head()
```

```python
ratings_dict = {'itemID': df_['JokeId'],
               'userID': df_['variable'],
               'rating': df_['rating']}
df = pd.DataFrame(ratings_dict)
df.head()
```
