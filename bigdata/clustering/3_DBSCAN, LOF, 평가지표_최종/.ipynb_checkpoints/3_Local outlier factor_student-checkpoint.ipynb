```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
```

```python
# LOF 알고리즘 연습을 위한 예제 데이터 생성
from sklearn.datasets import make_blobs

np.random.seed(1)  # for reproducibility
x, _ = make_blobs(n_samples=200, n_features=2,
                  centers=1, 
                  cluster_std=0.3, 
                  center_box=(10, 10))
```

```python
?make_blobs
```

```python
# plot example data
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```python
np.cov(x, rowvar = False)
```

## Local Outlier Factor(LOF)

```python
?LocalOutlierFactor
```

### 유클리디안 거리와 마할라노비스 거리

<img src="그림파일/mahalanobis_example.png" width="300" height="300">

```python
# Declare local outlier factor and get the result
lof = LocalOutlierFactor(n_neighbors=20, 
                         contamination=.1, 
                        metric='mahalanobis',
                        metric_params={'V': np.cov(x, rowvar=False)})
lof_result = lof.fit_predict(x)
```

```python
lof_result
```

```python
lof.negative_outlier_factor_
```

```python
lof.negative_outlier_factor_[lof_result == -1]
```

```python
# 군집화 결과 plotting
unique_labels = np.unique(lof_result)  # 군집 종류

for i in unique_labels:  # 각 군집에 대해
    cluster_instance_mask = (lof_result == i)
    
    x_cluster_i = x[cluster_instance_mask, :]  # 해당 군집에 해당하는 인스턴스
    
    plt.scatter(x_cluster_i[:, 0], x_cluster_i[:, 1], label='cluster ' + str(i))  # 1번째, 2번째 변수를 이용해 plotting

plt.title('example LOF result')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()
```

## 실습 내용

* 아래 생성된 data set을 통해 Local Outlier Factor를 측정하는 모형 만들기


* 하나의 모형은 마할라노비스 거리로, 하나의 모형은 유클리디언 거리로 모형 생성할 것


* contamination은 0.03과 0.1로 각각 변경해 생성해볼 것

```python
?make_blobs
```

```python
np.random.seed(1)  # for reproducibility
x, _ = make_blobs(n_samples=800, n_features=2,
                  centers=4, 
                  cluster_std= 1, 
                  center_box=(-20, 20))
```

```python
# plot example data
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```python
# Declare local outlier factor and get the result
lof = LocalOutlierFactor(n_neighbors=20, 
                         contamination=.03, 
                        metric='mahalanobis',
                        metric_params={'V': np.cov(x, rowvar=False)})
lof_result = lof.fit_predict(x)

# Declare local outlier factor and get the result
lof_e = LocalOutlierFactor(n_neighbors=20, 
                         contamination=.03, 
                        metric='euclidean')
lof_e_result = lof_e.fit_predict(x)

# 군집화 결과 plotting
unique_labels = np.unique(lof_result)  # 군집 종류

for i in unique_labels:  # 각 군집에 대해
    cluster_instance_mask = (lof_result == i)
    
    x_cluster_i = x[cluster_instance_mask, :]  # 해당 군집에 해당하는 인스턴스
    
    plt.scatter(x_cluster_i[:, 0], x_cluster_i[:, 1], label='cluster ' + str(i))  # 1번째, 2번째 변수를 이용해 plotting

plt.title('example LOF result')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()


# 군집화 결과 plotting
unique_labels = np.unique(lof_e_result)  # 군집 종류

for i in unique_labels:  # 각 군집에 대해
    cluster_instance_mask = (lof_e_result == i)
    
    x_cluster_i = x[cluster_instance_mask, :]  # 해당 군집에 해당하는 인스턴스
    
    plt.scatter(x_cluster_i[:, 0], x_cluster_i[:, 1], label='cluster ' + str(i))  # 1번째, 2번째 변수를 이용해 plotting

plt.title('example LOF result')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()
```

```python
# Declare local outlier factor and get the result
lof = LocalOutlierFactor(n_neighbors=20, 
                         contamination=.1, 
                        metric='mahalanobis',
                        metric_params={'V': np.cov(x, rowvar=False)})
lof_result = lof.fit_predict(x)

# Declare local outlier factor and get the result
lof_e = LocalOutlierFactor(n_neighbors=20, 
                         contamination=.1, 
                        metric='euclidean')
lof_e_result = lof_e.fit_predict(x)

# 군집화 결과 plotting
unique_labels = np.unique(lof_result)  # 군집 종류

for i in unique_labels:  # 각 군집에 대해
    cluster_instance_mask = (lof_result == i)
    
    x_cluster_i = x[cluster_instance_mask, :]  # 해당 군집에 해당하는 인스턴스
    
    plt.scatter(x_cluster_i[:, 0], x_cluster_i[:, 1], label='cluster ' + str(i))  # 1번째, 2번째 변수를 이용해 plotting

plt.title('example LOF result')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()


# 군집화 결과 plotting
unique_labels = np.unique(lof_e_result)  # 군집 종류

for i in unique_labels:  # 각 군집에 대해
    cluster_instance_mask = (lof_e_result == i)
    
    x_cluster_i = x[cluster_instance_mask, :]  # 해당 군집에 해당하는 인스턴스
    
    plt.scatter(x_cluster_i[:, 0], x_cluster_i[:, 1], label='cluster ' + str(i))  # 1번째, 2번째 변수를 이용해 plotting

plt.title('example LOF result')
plt.xlabel('X0')
plt.ylabel('X1')
plt.legend()
plt.show()
```
