### #3. Numeric 변수 전처리

```python
apt_list.isnull().sum(axis = 0) # 각 변수들의 NaN값 체크
```

```python
Num_idx1 = apt_list.area > 0 # 면적 outlier에 대한 idx 제거
Num_idx2 = apt_list.room > 0  # 방 갯수 outlier에 대한 idx 제거
Num_idx3 = apt_list.용적률.isnull() | (apt_list.용적률 < 1500) # 용적률 outlier에 대한 idx 제거 + NA값이 제거되는 것 방지
Num_idx4 = apt_list.건폐율.isnull() | (apt_list.건폐율 < 100) # 건폐율 outlier에 대한 idx 제거 + NA값이 제거되는 것 방지

Num_idx = Num_idx1 & Num_idx2 & Num_idx3 & Num_idx4
apt_list2 = apt_list.loc[Num_idx, :]
```

```python
print('아웃라이어 제거 후 해당 갯수:', Num_idx1.sum(), Num_idx2.sum(), Num_idx3.sum(), Num_idx4.sum())
print('최종 해당 갯수:', Num_idx.sum())
```

```python
print('개별냉난방 해당 갯수:', Chr_idx1.sum())
```

```python
# 데이터가 올바로 바뀌었는지 확인
Chr_idx2 = apt_list2.heat == '개별냉난방'
print('(수정 후) 개별냉난방 해당 갯수:', Chr_idx2.sum())
```

# 2. 국토교통부 실거래가 데이터

```python
trd_list = pd.read_csv('./Dataset/서울특별시_매매가_info.csv', encoding = 'cp949')
trd_list
```

```python
trd_list.info(), trd_list.describe()
```

### #1. 데이터 병합

* 아파트이름+전용면적 크기로 join key를 쓰면 안되는 이유

```python
# 중복된 아파트이름+전용면적 아파트 check
dup_list = apt_list.loc[:, ['아파트이름', 'area']].duplicated(keep = False)
apt_list.loc[dup_list, :].sort_values(by = ['아파트이름', 'area']) 
```

* 네이버 데이터에 대해 주소+전용면적 칼럼 생성

```python
apt_list2.loc[:, 'join_key'] = apt_list2.LotAdr+' '+ apt_list2.area.apply(lambda x: str(x))
apt_list2.join_key
```

* 국토교통부 데이터에 대해 주소+전용면적 칼럼 생성

```python
trd_list2 = trd_list
trd_list2.loc[:, 'join_key'] = trd_list2.시군구+' '+trd_list2.번지+' '+trd_list2.area.apply(lambda x: str(x))
trd_list2.join_key
```

```python
trd_list2.join_key = trd_list2.join_key.apply(
    lambda x: str(x).replace('서울특별시', '서울시')) # join key 값 형식을 맞추기 위해 변경
```

* 데이터 병합하기
> 주의: trd_list2를 기준으로 left_join할 것!

```python
merged_data = trd_list2.merge(apt_list2, how='left', on='join_key')
merged_data.info()
```

### #2. 데이터 Impute

```python
idx_nanroom = merged_data.room.isnull() # room 변수에서 null인 idx 찾기
idx_nantol = merged_data.toilet.isnull() # toilet 변수에서 null인 idx 찾기

print('room 변수에서 결측치 수:', idx_nanroom.sum())
print('toilet 변수에서 결측치 수:', idx_nantol.sum())
```

```python
merged_data.toilet
```

```python
# room 의 평균값과 toilet의 mode 값 확인
merged_data.room.mean(), merged_data.toilet.mode()[0]
```

```python
merged_data1 = merged_data
merged_data1.loc[idx_nanroom,"room"] = merged_data.room.mean() # ??에 알맞은 값 채워넣기
merged_data1.loc[idx_nantol,"toilet"] = merged_data.toilet.mode()[0] # ??에 알맞은 값 채워넣기
```

```python
# null 데이터가 없어졌는지 확인
merged_data1.room.isnull().sum(), merged_data1.toilet.isnull().sum()
```
