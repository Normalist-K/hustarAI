## Settings

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
```

## Problem 1 

```python
data = np.load('./data_files/data.npy')
data_labels = np.load('./data_files/data_labels.npy')
```

### (1) Print the data shape 

```python
print('Fashion mnist shape: ', data.shape)
print('Fashion mnist label shape: ', data_labels.shape)
```

```python
target_dict = {
 0: 'T-shirt/top',
 1: 'Trouser',
 2: 'Pullover',
 3: 'Dress',
 4: 'Coat',
 5: 'Sandal',
 6: 'Shirt',
 7: 'Sneaker',
 8: 'Bag',
 9: 'Ankle boot',
}
```

```python
plt.figure(figsize=(10,10))
for i in range(0,20):
    plt.subplot(5,5, i+1)
    plt.imshow(data[i], cmap=plt.cm.binary)
    plt.title(target_dict[(data_labels[i])])
    plt.xticks([])
    plt.yticks([])
```

```python
plt.figure()
plt.imshow(data[50], cmap=plt.cm.binary)
plt.colorbar()
plt.grid(False)
plt.show()
```

### (2) Plot random images for selected class

```python
selected_idx = np.where(data_labels == 1)[0]
plt.figure(figsize=(10,10))
for i in range(0,20):
    plt.subplot(5,5, i+1)
    plt.imshow(data[selected_idx[i]], cmap=plt.cm.binary)
    plt.title(target_dict[data_labels[selected_idx][0]])
    plt.xticks([])
    plt.yticks([])
```

### (3) Split train/test data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data,
                                                    data_labels,
                                                    test_size=0.3,
                                                    random_state=1)
print("Number of train data: {}".format(y_train.shape[0]))
print("Number of test data: {}".format(y_test.shape[0]))
```

## Problem 2

### (1) Make an ANN model

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape = (28, 28)),
    tf.keras.layers.Dense(units = 100, activation = 'relu'),
    tf.keras.layers.Dense(units = 50, activation = 'relu'),
    tf.keras.layers.Dense(units = 10, activation = 'softmax'),
])
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=50, batch_size = 256)
```

```python
#Training loss
loss = history.history['loss']
epochs = range(1, len(loss)+1)

plt.figure(figsize=(10,8))
plt.plot(epochs, loss, 'b-', label='Training Loss')
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

### (4) Evaluate the test data : loss, accuracy

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
```

```python
print(f"Fashion items test loss : {test_loss:.3f}")
print(f"Fashion items test accuracy : {test_acc:.3f}")
```

## Problem 3

```python
predictions = model.predict(x_test)
```

## (1) Visualize predicted values

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

def plot_image(i, predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array[i], true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color, size=15)
```

```python
plot_image(3, predictions, y_test, x_test)
```
