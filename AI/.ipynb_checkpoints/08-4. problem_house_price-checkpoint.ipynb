```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import time
import os
```

```python
train = pd.read_csv('./project/data_files/HousePrice_train.csv')
train
```

```python
print('Shape of the train data with all features:', train.shape)
train = train.select_dtypes(exclude = ['object'])
print("")
print('Shape of the train data with numerical features:', train.shape)
```

```python
train.drop('Id', axis = 1, inplace = True)
train.fillna(0, inplace = True)
```

```python
print("")
print('List of features contained our dataset:', list(train.columns))
print("")
print('Shape of the test data with numerical features:', train.shape)
```

```python
from sklearn.ensemble import IsolationForest

clf = IsolationForest(max_samples = 100, random_state = 42)
clf.fit(train)
y_noano = clf.predict(train)
y_noano = pd.DataFrame(y_noano, columns = ['Top'])
y_noano[y_noano['Top'] == 1].index.values

train = train.iloc[y_noano[y_noano['Top'] == 1].index.values]
train.reset_index(drop = True, inplace = True)
print("Number of Outliers:", y_noano[y_noano['Top'] == -1].shape[0])
print("Number of rows without outliers:", train.shape[0])
```

```python
train.head()
```

```python
train = np.asarray(train)
train.shape
```

```python
x_norm = train.copy()
x_norm.shape
```

```python
x_min = np.min(train, axis = 0)
x_min
```

```python
x_norm = np.subtract(x_norm, x_min)
```

```python
x_norm
```

```python
train
```

```python
x_max = np.max(train, axis = 0)
x_norm /= x_max
```

```python
x_max
```

```python
x_norm
```

```python
np.max(x_norm)
```

```python
x_norm.shape
```

```python
train_x_wo = train[:, :36]
train_y_wo = train[:, 36].reshape(-1,1)

train_x_norm = x_norm[:, :36]
train_y_norm = x_norm[:, 36].reshape(-1,1)

print('train_x_without normalization:', train_x_wo.shape)
print('train_y_without normalization:', train_y_wo.shape)
print('train_x_normalization:', train_x_norm.shape)
print('train_y_normalization:', train_y_norm.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(train_x_wo, train_y_wo)
train_X_n, test_X_n, train_Y_n, test_Y_n = train_test_split(train_x_norm, train_y_norm)
```

```python
import os
S = 0
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"]=str(S)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units = 200, input_dim = 36, activation = 'relu'),
    tf.keras.layers.Dense(units = 100, activation = 'relu'),
    tf.keras.layers.Dense(units = 50, activation = 'relu'),
    tf.keras.layers.Dense(units = 25, activation = 'relu'),
    tf.keras.layers.Dense(units = 1)
])
```

```python
model.compile(optimizer = 'adam',
              loss = 'mse',
              metrics = ['mae'])
#lr: 0.001 (default)
```

```python
model.summary()
```

```python
history = model.fit(train_X, train_Y, epochs = 500, batch_size=50)
```

```python
model_norm = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units = 200, input_dim = 36, activation = 'relu'),
    tf.keras.layers.Dense(units = 100, activation = 'relu'),
    tf.keras.layers.Dense(units = 50, activation = 'relu'),
    tf.keras.layers.Dense(units = 25, activation = 'relu'),
    tf.keras.layers.Dense(units = 1)
])
```

```python
model_norm.compile(optimizer = 'adam',
              loss = 'mse',
              metrics = ['mae'])
#lr: 0.001 (default)
```

```python
model_norm.summary()
```

```python
history_norm = model_norm.fit(train_X_n, train_Y_n, epochs = 500, batch_size=50)
```

```python
loss = history.history['loss']

epochs = range(1, len(loss)+1)

plt.figure(figsize=(10,8))
plt.plot(epochs, loss, 'b-', label='Training loss')
plt.title('w/o norm Training loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
loss = history_norm.history['loss']

epochs = range(1, len(loss)+1)

plt.figure(figsize=(10,8))
plt.plot(epochs, loss, 'b-', label='Training loss')
plt.title('w/o norm Training loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
with_y_pred = model_norm.predict(test_X_n)
```

```python
with_y_pred
```

```python
matplotlib.rc('xtick', labelsize=30) 
matplotlib.rc('ytick', labelsize=30) 

plt.figure(figsize=(10,8))
plt.style.use('default')
plt.plot(with_y_pred, test_Y_n, 'ro')
plt.xlabel('Predictions', fontsize = 15)
plt.ylabel('Reality', fontsize = 15)
plt.title('Predictions x Reality on dataset Test', fontsize = 20)

x1p = np.arange(0, 1, 0.001).reshape(-1, 1)
x2p = x1p
plt.plot(x1p, x2p, 'bo', markersize=2)

plt.show()
```
