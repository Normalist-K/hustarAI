```python
import numpy as np
import matplotlib.pyplot as plt

# 10 data points
n = 10
x = np.linspace(-4.5, 4.5, 10).reshape(-1, 1)
y = np.array([0.9819, 0.7973, 1.9737, 0.1838, 1.3180, -0.8361, -0.6591, -2.4701, -2.8122, -6.2512]).reshape(-1, 1)

plt.figure(figsize = (10, 8))
plt.plot(x, y, 'o', label = 'Data')
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.grid(alpha = 0.3)
plt.show()
```

```python
A = np.hstack([x**0, x])
A = np.asmatrix(A)

theta = (A.T*A).I * A.T * y
print('theta:\n', theta)
```

```python
# to plot
xp = np.arange(-4.5, 4.5, 0.01).reshape(-1, 1)
yp = theta[0, 0] + theta[1, 0] * xp

plt.figure(figsize = (10, 8))
plt.plot(x, y, 'o', label = 'Data')
plt.plot(xp[:,0], yp[:,0], linewidth = 2, label = 'Linear')
plt.title('Linear Regression', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.legend(fontsize = 15)
plt.grid(alpha = 0.3)
plt.show()
```

# 2. Regularization

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()

d = 10
reg.fit(np.hstack([x**i for i in range(d)]), y)

xp = np.arange(np.min(x), np.max(x), 0.01).reshape([-1, 1])
basis = np.hstack([xp**i for i in range(d)])

plt.figure(figsize = (10, 8))
plt.plot(x, y, 'o', label = 'Data')
plt.plot(xp, reg.predict(basis), label = 'Overfitted')
plt.title('(Overfitted) Regression', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.axis([-5, 5, -12, 6])
plt.legend(fontsize = 15)
plt.grid(alpha = 0.3)
plt.show()
```

```python
# ridge regression 
from sklearn import linear_model

reg = linear_model.Ridge()

d = 10
reg.fit(np.hstack([x**i for i in range(d)]), y)

xp = np.arange(np.min(x), np.max(x), 0.01).reshape([-1, 1])
basis = np.hstack([xp**i for i in range(d)])

plt.figure(figsize = (10, 8))
plt.plot(x, y, 'o', label = 'Data')
plt.plot(xp, reg.predict(basis), label = 'Ridge')
plt.title('Ridge Regularization (L2)', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.axis([-5, 5, -12, 6])
plt.legend(fontsize = 15)
plt.grid(alpha = 0.3)
plt.show()
```

```python
# Regulization (= ridge nonlinear regression) encourages small weights, but not exactly 0

plt.figure(figsize = (10, 8))
plt.title(r'Ridge: magnitude of $\theta$', fontsize = 15)
plt.xlabel(r'$\theta$', fontsize = 15)
plt.ylabel('magnitude', fontsize = 15)
plt.stem(np.arange(1,11), np.ravel(reg.coef_))
plt.xlim([0.5, 10.5])
# plt.ylim([-10, 10])
plt.grid(alpha = 0.3)
plt.show()
```

```python
lamb = np.arange(0,3,0.01)

theta_record = []
d = 10

for k in lamb:
    reg = linear_model.Ridge(alpha = k)
    reg.fit(np.hstack([x**i for i in range(d)]), y)
    theta_record.append(np.hstack([np.ravel(reg.coef_), reg.intercept_]))

plt.figure(figsize = (10, 8))
plt.plot(lamb, theta_record, linewidth = 1)
plt.title('Ridge coefficients as a function of regularization', fontsize = 15)
plt.xlabel('$\lambda$', fontsize = 15)
plt.ylabel(r'weight $\theta$', fontsize = 15)
plt.show()
```

```python
# LASSO regression 

from sklearn import linear_model

reg = linear_model.Lasso()

d = 10
reg.fit(np.hstack([x**i for i in range(d)]), y)

xp = np.arange(np.min(x), np.max(x), 0.01).reshape([-1, 1])
basis = np.hstack([xp**i for i in range(d)])

plt.figure(figsize = (10, 8))
plt.title('LASSO Regularization (L1)', fontsize = 15)
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.plot(x, y, 'o', label = 'Data')
plt.plot(xp, reg.predict(basis), label = 'LASSO')
plt.axis([-5, 5, -12, 6])
plt.legend(fontsize = 15)
plt.grid(alpha = 0.3)
plt.show()
```

```python
# Regulization (= Lasso nonlinear regression) encourages zero weights

plt.figure(figsize = (10, 8))
plt.title(r'LASSO: magnitude of $\theta$', fontsize = 15)
plt.xlabel(r'$\theta$', fontsize = 15)
plt.ylabel('magnitude', fontsize = 15)
plt.stem(np.arange(1,11), np.ravel(reg.coef_))

plt.xlim([0.5, 10.5])
plt.ylim([-0.5,0.5])
plt.grid(alpha = 0.3)
plt.show()
```

# 4. L2 and L1 Regularizers

## 4.1 L2

```python
x = np.arange(-4,4,0.1)

k = 4
y = (x-1)**2 + 1/6*(x-1) + k*x**2

x_star = x[np.argmin(y)]
print(x_star)
```

```python
plt.plot(x, y, 'g', linewidth = 2.5)
plt.axvline(x = x_star, color = 'k', linewidth = 1, linestyle = '--')
plt.ylim([0,10])
plt.show()
```

```python
for k in [0,1,2,4]:
    y = (x-1)**2 + 1/6*(x-1) + k*x**2
    x_star = x[np.argmin(y)]
    
    plt.plot(x,y, 'g', linewidth = 2.5)
    plt.axvline(x = x_star, color = 'k', linewidth = 1, linestyle = '--')
    plt.ylim([0,10])
    plt.title('Ridge: k = {}'.format(k))
    plt.show()
```

## L1

```python
x = np.arange(-4,4,0.1)
k = 2
y = (x-1)**2 + 1/6*(x-1) + k*abs(x)

x_star = x[np.argmin(y)]
print(x_star)
```

```python
plt.plot(x, y, 'g', linewidth = 2.5)
plt.axvline(x = x_star, color = 'k', linewidth = 1, linestyle = '--')
plt.ylim([0,10])
plt.show()
```

```python
for k in [0,1,2]:
    y = (x-1)**2 + 1/6*(x-1) + k*abs(x)
    x_star = x[np.argmin(y)]
    
    plt.plot(x,y, 'g', linewidth = 2.5)
    plt.axvline(x = x_star, color = 'k', linewidth = 1, linestyle = '--')
    plt.ylim([0,10])
    plt.title('LASSO: k = {}'.format(k))
    plt.show()
```
