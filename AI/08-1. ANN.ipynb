```python
mnist = tf.keras.datasets.mnist

(train_x, train_y), (test_x, test_y) = mnist.load_data()

train_x, test_x = train_x/255.0, test_x/255.0
```

```python
print ("The training data set is:\n")
print (train_x.shape)
print (train_y.shape)
```

```python
print ("The test data set is:")
print (test_x.shape)
print (test_y.shape)
```

```python
print('label :', train_y[5])

plt.figure(figsize = (6,6))
plt.imshow(train_x[5], 'gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# Make model
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape = (28, 28)),
    tf.keras.layers.Dense(units = 100, activation = 'relu'),
    tf.keras.layers.Dense(units = 10, activation = 'softmax')
])
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
loss = model.fit(train_x, train_y, epochs = 8)
```

```python
test_loss, test_acc = model.evaluate(test_x, test_y)
```

```python
test_img = test_x[np.random.choice(test_x.shape[0], 1)]

predict = model.predict_on_batch(test_img)
mypred = np.argmax(predict, axis = 1)

plt.figure(figsize = (12,5))

plt.subplot(1,2,1)
plt.imshow(test_img.reshape(28, 28), 'gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.stem(predict[0])
plt.show()

print('Prediction : {}'.format(mypred[0]))
```
