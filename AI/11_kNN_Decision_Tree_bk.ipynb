# 2. Setting

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing as pp
from sklearn.cluster import KMeans

sns.set()
%matplotlib inline
```

# 3. Data

## 3.1 Data_Load

```python
iris = pd.read_csv("./Iris.csv")
iris.shape
```

## 3.2 Data_Print

```python
iris.head()
```

## 3.3 Data_Drop

```python
#Drop : Id
X = iris.iloc[:,1:5]
```

```python
X.shape
```

```python
X = X.to_numpy()
X
```

## 3.4 Data_Labeling

```python
# Data_value : Species
iris.loc[:,'Species'].to_numpy()
```

- **0 = setosa, 1 = versicolor, 2 = virginica**

```python
y = pd.Categorical(iris['Species']).codes
```

```python
Species = y.reshape(-1,1)
```

```python
Species.shape
```

```python
Species
```

## 3.5 Data_Combined

```python
# Data_Combined
data = np.hstack([X, Species])
```

```python
data
```

```python
# DataFrame : Data_Combined
df = pd.DataFrame(data, columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])
```

```python
df
```

# 4. Select Variables

## 4.1 Select Variables_Correlation

```python
# Correlation
corr_mat = df.corr()
corr_mat
```

### 4.1.1 Correlation_Visualization

```python
# Visualize : Correlation (Heatmap)
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(corr_mat, cmap='Blues', annot=True, annot_kws={'size':15}, linewidths=.5, fmt='.2f')
```

# 5. Selected_Variables & Target Data

```python
# Select x/y data
data_x_y = df.loc[:,["SepalLengthCm", "PetalLengthCm", "PetalWidthCm", "Species"]]
data_x_y
```

## 5.1 Count_Target

```python
target = data_x_y["Species"].astype('int')
```

```python
print("Real Values : ")
real_unique_class , real_unique_class_counts = np.unique(target, return_counts=True)
for x , y in zip(real_unique_class, real_unique_class_counts):
    print("The number of observations assigned class",x,"is",y)
```

# 6. Split_Train/Test dataset

```python
from sklearn.model_selection import train_test_split
```

```python
# split Train/Test
#test_size = 0.3
Selected_variables = data_x_y[["SepalLengthCm", "PetalLengthCm", "PetalWidthCm"]]
x_train, x_test, y_train, y_test = train_test_split(Selected_variables,
                                                   target, test_size = 0.3,
                                                   random_state = 1)
print(f'# of train data: {y_train.shape[0]}')
print(f'# of test data: {y_test.shape[0]}')
```

# 7. Data_Visualization

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
train_data = pd.concat([x_train, y_train], axis=1)

plt.style.use('default')
fig=plt.figure(figsize=(8,6))
ax=Axes3D(fig)
ax.scatter(train_data['SepalLengthCm'][train_data['Species']==0],
           train_data['PetalLengthCm'][train_data['Species']==0],
           train_data['PetalWidthCm'][train_data['Species']==0],
           s = 30, c = 'green', label = 'Setosa')
ax.scatter(train_data['SepalLengthCm'][train_data['Species']==1],
           train_data['PetalLengthCm'][train_data['Species']==1],
           train_data['PetalWidthCm'][train_data['Species']==1],
           s = 30, c = 'blue', label = 'Versicolor')
ax.scatter(train_data['SepalLengthCm'][train_data['Species']==2],
           train_data['PetalLengthCm'][train_data['Species']==2],
           train_data['PetalWidthCm'][train_data['Species']==2],
           s = 30, c = 'red', label = 'Virginica')
ax.set_title('Train data visualization', size = 13)
ax.set_xlabel('Sepal_Length  (cm)', size = 10)
ax.set_ylabel('Petal_Length (cm)', size = 10)
ax.set_zlabel('Petal_Width  (cm)', size = 10)
plt.legend(fontsize=10)
plt.show()
```

```python
# Visualization test data 3D
test_data = pd.concat([x_test, y_test], axis=1)

plt.style.use('default')
fig=plt.figure(figsize=(8,6))
ax=Axes3D(fig)
ax.scatter(test_data['SepalLengthCm'][test_data['Species']==0],
           test_data['PetalLengthCm'][test_data['Species']==0],
           test_data['PetalWidthCm'][test_data['Species']==0],
           s = 30, c = 'green', label = 'Setosa')
ax.scatter(test_data['SepalLengthCm'][test_data['Species']==1],
           test_data['PetalLengthCm'][test_data['Species']==1],
           test_data['PetalWidthCm'][test_data['Species']==1],
           s = 30, c = 'blue', label = 'Versicolor')
ax.scatter(test_data['SepalLengthCm'][test_data['Species']==2],
           test_data['PetalLengthCm'][test_data['Species']==2],
           test_data['PetalWidthCm'][test_data['Species']==2],
           s = 30, c = 'red', label = 'Virginica')
ax.set_title('test data visualization', size = 13)
ax.set_xlabel('Sepal_Length  (cm)', size = 10)
ax.set_ylabel('Petal_Length (cm)', size = 10)
ax.set_zlabel('Petal_Width  (cm)', size = 10)
plt.legend(fontsize=10)
plt.show()
```

# 8. Classification

## 8.1 KNN_neighbors

**neighbors : 3**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

**training & test**

```python
# kNN Classifier n = 3
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
# kNN train
clf.fit(x_train, y_train)
```

```python
#Predict
y_pred = clf.predict(x_test)
```

**Accuracy**

```python
knn_3_acc=metrics.accuracy_score(y_test,y_pred)
print('knn_3_neighbors accuracy :', np.round(knn_3_acc*100,1),'%')
```

### 8.1.1 Confusion_matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
#Plot 3-Class Confusion matrix
knn_3_cm = confusion_matrix(y_test, y_pred)
conf_matrix = pd.DataFrame(knn_3_cm,
                           columns=['Predict:0(Setosa)', 'Predict:1(Versicolor)', 'Predict:2(Virginica)'],
                           index=['Actual:0(Setosa)', 'Actual:1(Versicolor)', 'Actual:2(Virginica)'])
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(conf_matrix, cmap='Blues', annot=True, annot_kws={'size':15}, linewidths=.5, fmt='.2f')
```

```python
knn_3_cm_percent = knn_3_cm.astype('float')/knn_3_cm.sum(axis=1)[:, np.newaxis]
conf_matrix_percent = pd.DataFrame(knn_3_cm_percent,
                           columns=['Predict:0(Setosa)', 'Predict:1(Versicolor)', 'Predict:2(Virginica)'],
                           index=['Actual:0(Setosa)', 'Actual:1(Versicolor)', 'Actual:2(Virginica)'])

#Plot 3-Class Confusion matrix(%)
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(conf_matrix_percent, cmap='Blues', annot=True, annot_kws={'size':15}, linewidths=.5, fmt='.2f')
plt.xticks(size=10)
plt.yticks(size=10, rotation=0)
```

### 8.1.2 Predicted values_Visualization

```python
x_test_indices = x_test.index
y_pred_data = y_pred.reshape(-1,1)
y_pred_data = pd.DataFrame(y_pred_data, columns = ['pred_y'], index=x_test_indices)
```

```python
# plot
test_data = pd.concat([x_test, y_pred_data], axis=1)

plt.style.use('default')
fig=plt.figure(figsize=(8,6))
ax=Axes3D(fig)
ax.scatter(test_data['SepalLengthCm'][test_data['pred_y']==0],
           test_data['PetalLengthCm'][test_data['pred_y']==0],
           test_data['PetalWidthCm'][test_data['pred_y']==0],
           s = 30, c = 'green', label = 'Setosa')
ax.scatter(test_data['SepalLengthCm'][test_data['pred_y']==1],
           test_data['PetalLengthCm'][test_data['pred_y']==1],
           test_data['PetalWidthCm'][test_data['pred_y']==1],
           s = 30, c = 'blue', label = 'Versicolor')
ax.scatter(test_data['SepalLengthCm'][test_data['pred_y']==2],
           test_data['PetalLengthCm'][test_data['pred_y']==2],
           test_data['PetalWidthCm'][test_data['pred_y']==2],
           s = 30, c = 'red', label = 'Virginica')
ax.set_title('test data visualization', size = 13)
ax.set_xlabel('Sepal_Length  (cm)', size = 10)
ax.set_ylabel('Petal_Length (cm)', size = 10)
ax.set_zlabel('Petal_Width  (cm)', size = 10)
plt.legend(fontsize=10)
plt.show()
```

### 8.1.3 Varing number of neighbors

```python
neighbors = np.arange(1,9)
test_accuracy = np.empty(len(neighbors))
```

```python
# kNN Accuracy (neighbors = 1 ~ 9)
# plot
for neighbor in neighbors:
    clf = KNeighborsClassifier(n_neighbors=neighbor)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    knn_3_acc=metrics.accuracy_score(y_test,y_pred)
    test_accuracy[neighbor-1] = knn_3_acc
    
plt.figure(figsize=(5, 4))
plt.title(label='k-NN Varying number of neighbors')
plt.plot(neighbors, test_accuracy, '--o', label='Testing Accuracy', markersize=3)
plt.legend()
plt.xlabel('# of neighbors')
plt.ylabel('Accuracy')
plt.xticks(size=10)
plt.yticks(size=10, rotation=0)
plt.show()
```
